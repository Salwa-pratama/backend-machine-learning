```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score, classification_report,precision_score,recall_score,f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Dataset/Healthcare-Diabetes.csv")
```

```python
x = df.drop(["Outcome", "Id"], axis=1)
y = df["Outcome"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3,random_state=42, stratify=y
)
```

# Scaller

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
joblib.dump(scaler, "Scaler_nn.pkl")
```

```python
model_nn = MLPClassifier(
    hidden_layer_sizes=(16, 8),
    activation="relu",
    solver="adam",
    max_iter=500,
    random_state=42
)
```

```python
model_nn.fit(x_train_scaled, y_train)
y_pred = model_nn.predict(x_test_scaled)
```

```python
print("\nAccuracy:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

```

# Importing

```python
# Model
# joblib.dump(model_nn, "Model_nn.pkl")

# Evaluasi
precision = precision_score(y_test, y_pred, average="weighted")
recall = recall_score(y_test, y_pred, average="weighted")
f1 = f1_score(y_test, y_pred, average="weighted")
report = classification_report(y_test, y_pred, output_dict=True)
# joblib.dump({
#     "precision" : precision,
#     "recall" : recall,
#     "f1_score" : f1,
#     "report" : report
# }, "Evaluation_NN.pkl")
acc = accuracy_score(y_test,y_pred)
# joblib.dump(acc,"AccNN.pkl")

print("Accuracy:", acc)
print("Classification Report")
classification_report(y_test, y_pred)
```
